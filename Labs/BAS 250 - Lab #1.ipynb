```python
#creating second dataset using only specified columns.  We create a copt using .copy() as we want to create a copy of the dataset
#not just a subset of the original dataset
data2 = data1[['Created Date', 'Agency', 'Agency Name', 'Complaint Type', 'Descriptor', 'Location Type',
               'Incident Zip', 'Borough', 'Latitude', 'Longitude']].copy()
```

```python
#reviewing top 5 rows of new dataset
data2.head()
```

```python
#renaming columns
data2.columns = ['Date', 'Agency', 'Agency_name', 'Complaint_type', 'Descriptor', 'Location_type', 'Zip', 'Borough', 
                'Latitude', 'Longitude']
```

```python
#creating new variable from Date that takes the Date variable and changes it to dtype of datetime
data2['Date2'] = pd.to_datetime(data2['Date'])

#we can then drop the original Date column
data2 = data2.drop(columns = 'Date')
```

```python
#reordering columns in dataset
data2 = data2[['Date2', 'Agency', 'Agency_name', 'Complaint_type', 'Descriptor', 'Location_type', 'Zip', 'Borough', 
                'Latitude', 'Longitude']].copy()
```
